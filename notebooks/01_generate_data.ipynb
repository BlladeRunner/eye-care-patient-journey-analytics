# Synthetic Data Generation — Eye Care Patient Journey

## Purpose
This notebook generates a **fully synthetic, privacy-safe dataset** that simulates the patient journey in eye care:
from clinical recommendation to purchase, retention, and downstream revenue.

The goal is to enable **end-to-end analytics** (funnel, cohorts, uplift) while ensuring:
- full reproducibility,
- no use of sensitive or proprietary patient data,
- realistic relational structure between entities.

## Data Model Overview

The synthetic dataset consists of multiple related tables:

- **patients** — demographic and market attributes
- **visits** — clinical visits and recommendations
- **recommendations** — recommended product tier and category
- **transactions** — purchases, revenue, channel
- **outcomes** — adoption and follow-up signals

Relationships:
- One patient → many visits
- One recommendation → optional purchase
- Purchases generate revenue and retention signals

Distributions and probabilities are intentionally simplified but calibrated
to produce realistic funnel drop-offs and cohort behavior.

## Reproducibility

Running this notebook from top to bottom will regenerate all raw CSV files
in `data/raw/`, allowing the entire project to be reproduced from scratch.

This approach mirrors real-world constraints in healthcare analytics,
where analysts often work with synthetic or anonymized datasets.
